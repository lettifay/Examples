```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
im = cv2.imread('./test.jpg')
```

```python
plt.imshow(im)
```

```python
# get haar xml from https://github.com/opencv/opencv/blob/master/data/haarcascades/haarcascade_frontalface_default.xml
haar_xml = './haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(haar_xml)

# convert to gray to accelerate
gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)

# detectMultiScale(img,scaleFactor=1.1,minNeighbors=3)
faces = face_cascade.detectMultiScale(gray,1.3,5)
print faces # [x,y,w,h]
```

```python
for (x,y,w,h) in faces:
    # rectangle(img,start_point,end_point,color,line_thickness)
    cv2.rectangle(im,(x,y),(x+w,y+h),(0,255,0),2) 
    plt.imshow(im)
```
